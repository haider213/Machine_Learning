We are going to write the very basics of neural networks using tensorflow and specifically keras

```python
import keras.layers
import pandas as pd
import numpy as np
import keras
from keras import Sequential
import tensorflow as tf
import matplotlib.pyplot as plt
```

Let us assume we have a data set that looks like this

```python
x=np.arange(-1,5)
```

```python
y=2*x-1
```

```python
y
```

```python
plt.plot(x,y)
```

We already know the relationship between x and y but let's suppose we did not, in the ML paradigm we give both the input and the output to the ML algorithm to figure out the relationship

Let us now define a linear model using Sequential from keras

```python
model=Sequential(keras.layers.Dense(units=1, input_shape=[1]))
```

Another way of defining the same thing is
a linear activation is characterised by the following $a=w*x+b$
In keras you use Dense to define a layer of connected neurons

```python
a=keras.layers.Dense(units=1,input_shape=[1],activation='linear')
```

```python
a.build(input_shape=[1])
```

```python
a
```

SGD is single gradient descent with momentum explained here: https://keras.io/api/optimizers/sgd/

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
model.summary()
```

```python
model.fit?
```

```python
model.fit(x,y,epochs=1000)
```

```python
model.get_weights()
```

```python
print(model.predict([10]))
```
